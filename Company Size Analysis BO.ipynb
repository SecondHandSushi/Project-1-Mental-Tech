```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from pprint import pprint
from pathlib import Path

# File locations
survey_2016 = "Resources/2016-cleaned.csv"

# Reading the data
clean_2016 = pd.read_csv(survey_2016)
```

```python
#Create DF for Comfort of MH with direct Supervisor vs Number of Employees
comp_size_summary = clean_2016[["comf_discuss_mh_disorder_cur_superv","num_employees"]]


comp_size_summary
```

```python
# Get counts in percentages by response
one_five = comp_size_summary.loc[comp_size_summary["num_employees"] == "1-5",:]
one_five_result = one_five[["comf_discuss_mh_disorder_cur_superv","num_employees"]]

one_five_count =one_five_result["comf_discuss_mh_disorder_cur_superv"].value_counts(normalize="True")
one_five_count 
```

```python
#Creat Pie Chart 1-5 Employees
plt.pie(one_five_count .values,labels=one_five_count.index.values, autopct="%1.1f%%")
plt.title("1-5 Employees")
plt.show()

```

```python
#Convert to Piviot Table and Plot
comp_size_summary_groupby.reset_index().pivot(columns="comf_discuss_mh_disorder_cur_superv", values=0, index="num_employees"
                                            ).plot.bar(xlabel="Company Size", ylabel="Response", title="Discussing Mental Health by Company Size")
plt.legend(title="Response")


```
