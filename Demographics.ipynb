```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from pathlib import Path

# File locations
survey_2016 = "Resources/2016-cleaned.csv"
combined = "Resources/combined_df.csv"

# Reading the data
df_2016 = pd.read_csv(survey_2016)
combined_df = pd.read_csv(combined)
```

```python
genders = ["Male", "Female", "Other"]
counts = df_2016['gender'].value_counts().to_list()
colors = ["lightblue", "pink", "lightcoral"]

plt.figure(figsize=(6, 6))

plt.pie(counts, labels=genders, autopct='%1.1f%%', colors=colors, \
        shadow=True, startangle=70)

plt.axis('equal')
plt.title('Gender Responses(2016)')
plt.savefig('Figures/demographic_gender_2016.png')
```

```python
ages = df_2016['age groups']
age_groups = ["18-24","25-34","35-44","45-54","55-64","65-74","75+"]
age_counts = ages.value_counts().to_list()

plt.figure(figsize=(10, 6))
bar_width = 0.4  

plt.bar(age_groups, age_counts, width=bar_width, label='Count')

plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.title('Age Responses (2016)')
plt.legend()
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
```
