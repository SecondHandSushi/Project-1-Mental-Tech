```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from pathlib import Path

# File locations
survey_2016 = "Resources/2016-cleaned.csv"

# Reading the data
df_2016 = pd.read_csv(survey_2016)
```

```python
def analysis_func(col1, col2, a1, a2, a3, q1):
    sample_df = df_2016[[col1, col2]]
    sample_df

    a1_counts = sample_df[sample_df[col1] == a1][col2].value_counts()
    a2_counts = sample_df[sample_df[col1] == a2][col2].value_counts()
    a3_counts = sample_df[sample_df[col1] == a3][col2].value_counts()

    categories = sample_df[col2].unique()

    positions = np.arange(len(categories))

    plt.figure(figsize=(10, 6))
    bar_width = 0.2 
    plt.bar(positions, a1_counts, width=bar_width, label=a1, align='center')
    plt.bar(positions + bar_width, a2_counts, width=bar_width, label=a2, align='center')
    plt.bar(positions + 2 * bar_width, a3_counts, width=bar_width, label=a3, align='center')

    plt.xticks(positions + bar_width, categories)

    # plt.xlabel('Would you feel comfortable discussing a mental health disorder with your direct supervisor(s)?')
    # plt.ylabel('Count')
    plt.title(q1)
    plt.legend()
    plt.legend(title='Do you work remote?')
    plt.savefig(f'Figures/{col1}_{col2}.png')
    plt.show()
```

```python
analysis_func('remote', 'obs_of_other_inhibit_discuss_curr', 'Always', 'Sometimes', 'Never', 'Have your observations of how another individual who discussed a mental health disorder made you less likely to reveal a mental health issue yourself in your current workplace?')
```
