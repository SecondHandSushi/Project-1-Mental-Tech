```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from pathlib import Path

# File locations
survey_2016 = "Resources/2016-cleaned.csv"

# Reading the data
df_2016 = pd.read_csv(survey_2016)
```

```python
# Remote Versus comf_discuss_mh_disorder_cur_superv
remote_df = df_2016[['remote', 'comf_discuss_mh_disorder_cur_superv']]
remote_df
```

```python
always_counts = remote_df[remote_df['remote'] == 'Always']['comf_discuss_mh_disorder_cur_superv'].value_counts()
sometimes_counts = remote_df[remote_df['remote'] == 'Sometimes']['comf_discuss_mh_disorder_cur_superv'].value_counts()
never_counts = remote_df[remote_df['remote'] == 'Never']['comf_discuss_mh_disorder_cur_superv'].value_counts()

categories = remote_df['comf_discuss_mh_disorder_cur_superv'].unique()

positions = np.arange(len(categories))

plt.figure(figsize=(10, 6))
bar_width = 0.2 
plt.bar(positions, always_counts, width=bar_width, label='Always', align='center')
plt.bar(positions + bar_width, sometimes_counts, width=bar_width, label='Sometimes', align='center')
plt.bar(positions + 2 * bar_width, never_counts, width=bar_width, label='Never', align='center')

plt.xticks(positions + bar_width, categories)

# plt.xlabel('Would you feel comfortable discussing a mental health disorder with your direct supervisor(s)?')
# plt.ylabel('Count')
plt.title('Would you feel comfortable discussing a mental health disorder with your direct supervisor(s)?')
plt.legend()
plt.legend(title='Do you work remotely?')
plt.savefig('Figures/remote_comf_sup.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

bar_width = 0.4 
plt.bar(positions, never_counts, width=bar_width, label='Never', align='center', color='#2ca02c')

plt.xticks(positions, categories)

plt.title('Would you feel comfortable discussing a mental health disorder with your direct supervisor(s)?')
plt.legend()
plt.legend(title='Do you work remotely?')
plt.savefig('Figures/never_remote_comf_sup.png')
plt.show()
```
