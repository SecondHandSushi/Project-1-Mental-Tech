```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from pathlib import Path

# File locations
survey_2016 = "Resources/2016-cleaned.csv"

# Reading the data
df_2016 = pd.read_csv(survey_2016)
```

```python
# Remote Versus comf_discuss_mh_disorder_cur_superv

# 
gender_vs_int_df = df_2016[['gender', 'bring_up_mh_issue_in_interview']]
gender_vs_int_df
```

```python
male_counts = gender_vs_int_df[gender_vs_int_df['gender'] == 'male']['bring_up_mh_issue_in_interview'].value_counts()
female_counts = gender_vs_int_df[gender_vs_int_df['gender'] == 'female']['bring_up_mh_issue_in_interview'].value_counts()
other_counts = gender_vs_int_df[gender_vs_int_df['gender'] == 'other']['bring_up_mh_issue_in_interview'].value_counts()

categories = gender_vs_int_df['bring_up_mh_issue_in_interview'].unique()

positions = np.arange(len(categories))

plt.figure(figsize=(10, 6))
bar_width = 0.2 
plt.bar(positions, male_counts, width=bar_width, label='Male', align='center')
plt.bar(positions + bar_width, female_counts, width=bar_width, label='Female', align='center')
plt.bar(positions + 2 * bar_width, other_counts, width=bar_width, label='Other', align='center')

plt.xticks(positions + bar_width, categories)

# plt.xlabel('Would you feel comfortable discussing a mental health disorder with your direct supervisor(s)?')
# plt.ylabel('Count')
plt.title('Would you bring up a mental health issue with a potential employer in an interview?')
plt.legend()
# plt.legend(title='')
# plt.savefig('Figures/remote_comf_sup.png')
plt.show()
```
